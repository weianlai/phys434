```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson
```

```python
# Part A
#
# The rate of cosmic ray background I have chosen is 3, which is plugged right into the
# poisson function. With increasing time of observation, the poisson parameter increases,
# since the variable is the product of the rate and the time interval. This poisson parameter,
# or lambda, as it gets larger, will peak later at a smaller probability. This behavior makes
# the curve more flattened as time interval is increased.

xaxis = np.arange(25);
rate = 3;

for x in range(1,5):
    dist = poisson(x * rate)
    plt.plot(xaxis, dist.pmf(xaxis), label = '%i days' %x);

plt.title('Poisson Distribution')
plt.xlabel('Days');
plt.ylabel('Probability');
plt.legend();
plt.show();
```

```python
# Part B
#
# The sum of the distribution should retain its Poisson behavior because the events are still
# occuring independently without effecting the next events. This follows the Central Limit Theorem,
# because as we put more days in, we increase the probability of occurance in later days, lowering
# the existing peak. As a result, the curve becomes more symmetrical at the mean, slowly becoming
# more like a normal distribution.

def poissonEq(mu, k):
    return np.exp(-mu) * mu**k / np.math.factorial(k)

for x in range(1,7):
    plt.scatter(x,poissonEq(rate,x))

plt.title('Poisson Distribution')
plt.xlabel('Days');
plt.ylabel('Probability');
plt.show();
```

```python
# Part C
#
# As we average more days, we can see that the curve increases, eventually getting very close to one
# when the days becomes very large. The behavior of the curve can be visualized by looking at the area
# of the original poisson ditribution curve. Most of the area is near the mean, or value of lambda,
# so the probability density adds up very quickly. At the skew, or the tail, there isn't much area, so
# there isn't much to add to the probability density.
#
# This is contrasting to the previous part, because Part B retains the behavior of the Poisson
# Distribution, while this problem shows an decreasing upward curve with limit at one. This also does
# not follow the Central Limit Theorem, as we can see that adding more independent events will never
# alter the graph to resemble anything like a Gaussian.

def aveDays(mu, days):
    count = 1;
    output = 0;
    
    for x in range(1,days + 1):
        output = output + poissonEq(mu, x);
    
    return output;

for x in range(1,11):
    plt.scatter(x,aveDays(3,x), label = '%i Average Day' %x)
    
plt.title('Evolution of Probability Distribution')
plt.xlabel('Average Days');
plt.ylabel('Probability');
plt.show();
```

```python
# Part D
#
# The average or mean of the Poisson Distribution is lambda, or at around the peak of the curve.
# The variance is also the same value as the lambda value. The sigma will be the squareroot of
# the variance, or the standard variation. We can obtain the answer by using python function, or
# take the root of the variance.

numberDays = 15; # number of days, N
source = 8; # gamma rays from source

result = poisson.std(numberDays * source)
print('Observation Sigma: %f' %result)
```

```python
# Part A
# 
from scipy.stats import rayleigh

# f(x) = x * exp(-x^2 / 2)

x = np.linspace(0,3,100)
x2 = np.linspace(0,10,100)
x3 = np.linspace(0,20,100)

rv = rayleigh()

plt.plot(x, rv.pdf(x), label = 'Interval = 3')
plt.plot(x, rv.pdf(x2), label = 'Interval = 10')
plt.plot(x, rv.pdf(x3), label = 'Interval = 20')

plt.title('Changes In Distribution')
plt.xlabel('Observing Interval')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```

```python
# Part B
#
# As the interval increases, the curve becomes more compact towards the y-axis.
# The peak shifts to the left, the width decreases, while the peak value stays the same.
# The tail on the right takes a more drastic change, as it quickly levels out at zero.
# With the tail fading, the curve resembles more like a Gaussian distribution.
```
